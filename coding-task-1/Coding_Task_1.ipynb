MENCARI NILAI TENGAH (MEDIAN)

```python
def cari_median(data, n, bulan, penjualan):
    index_tengah = n//2
    if n % 2 == 0:
        median1 = data[index_tengah]
        median2 = data[index_tengah - 1]
        median_value = (median1 + median2) / 2
        print("Median penjualan", penjualan, bulan, "=", median_value)
    else:
        median_value = data[index_tengah]
        print("Median penjualan", penjualan, bulan, "=", median_value)

#Mencari median penjualan A (Januari - Mei)
median_Jan_A = sorted(Jan_A)
cari_median(median_Jan_A, len(Jan_A), "Januari", "A")

median_Feb_A = sorted(Feb_A)
cari_median(median_Feb_A, len(Feb_A), "Februari", "A")

median_Mar_A = sorted(Mar_A)
cari_median(median_Mar_A, len(Mar_A), "Maret", "A")

median_Apr_A = sorted(Apr_A)
cari_median(median_Apr_A, len(Apr_A), "April", "A")

median_Mei_A = sorted(Mei_A)
cari_median(median_Mei_A, len(Mei_A), "Mei", "A")

median_penjualanA = sorted(penjualan_A)
cari_median(median_penjualanA, len(penjualan_A), "Januari - Mei", "A")
print()

#Mencari median penjualan B (Januari - Mei)
median_Jan_B = sorted(Jan_B)
cari_median(median_Jan_B, len(Jan_B), "Januari", "B")

median_Feb_B = sorted(Feb_B)
cari_median(median_Feb_B, len(Feb_B), "Februari", "B")

median_Mar_B = sorted(Mar_B)
cari_median(median_Mar_B, len(Mar_B), "Maret", "B")

median_Apr_B = sorted(Apr_B)
cari_median(median_Apr_B, len(Apr_B), "April", "B")

median_Mei_B = sorted(Mei_B)
cari_median(median_Mei_B, len(Mei_B), "Mei", "B")

median_penjualanB = sorted(penjualan_B)
cari_median(median_penjualanB, len(penjualan_B), "Januari - Mei", "B")
```

MENCARI MODUS (MODE)

```python
from collections import Counter
def mode(data, bulan, penjualan):
    counter = Counter(data)
    mode_data = dict(counter)
    max_counter = max(list(counter.values()))
    
    mode_value = []
    for k, v in mode_data.items():
        if v == max(list(counter.values())):
            mode_value.append(k)
    print("Modus penjualan", penjualan, bulan, "=", mode_value, "muncul", max_counter, "kali")

#Mencari modus penjualan A (Januari - Mei)
mode(sorted(Jan_A), "Januari", "A")

mode(sorted(Feb_A), "Februari", "A")

mode(sorted(Mar_A), "Maret", "A")

mode(sorted(Apr_A), "April", "A")

mode(sorted(Mei_A), "Mei", "A")

print()

#Mencari modus penjualan B (Januari - Mei)
mode(sorted(Jan_B), "Januari", "B")

mode(sorted(Feb_B), "Februari", "B")

mode(sorted(Mar_B), "Maret", "B")

mode(sorted(Apr_B), "April", "B")

mode(sorted(Mei_B), "Mei", "B")
```

MENCARI RANGE

```python
#Fungsi mencari Range (data terbesar - data terkecil)
def range(data, bulan, penjualan):
    nilai_tertinggi = max(data)
    nilai_terendah = min(data)
    data_range = nilai_tertinggi - nilai_terendah
    print("Range penjualan", penjualan, bulan, "=", data_range)
    
#Mencari Range penjualan A (Januari - Mei)
range(Jan_A, "Januari", "A")

range(Feb_A, "Februari", "A")

range(Mar_A, "Maret", "A")

range(Apr_A, "April", "A")

range(Mei_A, "Mei", "A")

print()

#Mencari Range penjualan B (Januari - Mei)
range(Jan_B, "Januari", "B")

range(Feb_B, "Februari", "B")

range(Mar_B, "Maret", "B")

range(Apr_B, "April", "B")

range(Mei_B, "Mei", "B")
```

MENCARI INTERQUARTIL RANGE (IQR)

```python
#Fungsi mencari IQR (Q3 - Q1)
def iqr(data, bulan, penjualan):
    Q1 = np.percentile(data, 25) # Kuartil pertama (25%)
    Q3 = np.percentile(data, 75) # Kuartil ketiga (75%)
    IQR = Q3 - Q1
    print("Interquartil Range penjualan", penjualan, bulan, "=", IQR)

#Mencari Interquartil Range penjualan A (Januari - Mei)
iqr(Jan_A, "Januari", "A")

iqr(Feb_A, "Februari", "A")

iqr(Mar_A, "Maret", "A")

iqr(Apr_A, "April", "A")

iqr(Mei_A, "Mei", "A")

print()

#Mencari Interquartil Range penjualan B (Januari - Mei)
iqr(Jan_B, "Januari", "B")

iqr(Feb_B, "Februari", "B")

iqr(Mar_B, "Maret", "B")

iqr(Apr_B, "April", "B")

iqr(Mei_B, "Mei", "B")
```

MENCARI VARIANSI

```python
#Fungsi mencari Variansi (jumlah setiap suku – rata-rata)^ 2 / n
def variance(data, mean, bulan, penjualan):
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append(squared_diff_per_x)
    varian = sum(squared_diff) / len(data)
    print(f"Variansi penjualan {penjualan} {bulan} = {varian:.2f}")

#Mencari Variansi penjualan A (Januari - Mei)
variance(Jan_A, mean_Jan_A, "Januari", "A")

variance(Feb_A, mean_Feb_A, "Februari", "A")

variance(Mar_A, mean_Mar_A, "Maret", "A")

variance(Apr_A, mean_Apr_A, "April", "A")

variance(Mei_A, mean_Mei_A, "Mei", "A")

print()

#Mencari Variansi penjualan B (Januari - Mei)
variance(Jan_B, mean_Jan_B, "Januari", "B")

variance(Feb_B, mean_Feb_B, "Februari", "B")

variance(Mar_B, mean_Mar_B, "Maret", "B")

variance(Apr_B, mean_Apr_B, "April", "B")

variance(Mei_B, mean_Mei_B, "Mei", "B")
```

STANDAR DEVIASI

```python
import math
def variance(data, mean):
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append(squared_diff_per_x)
    varian = sum(squared_diff) / len(data)
    return varian
    
def standard_deviation(data, mean, penjualan, bulan):
    varian = variance(data, mean)
    deviasi = math.sqrt(varian)
    print(f"Standard Deviasi penjualan {penjualan} {bulan} = {deviasi:.2f}")

#Mencari Standard Deviasi penjualan A (Januari - Mei)
standard_deviation(Jan_A, mean_Jan_A, "Januari", "A")

standard_deviation(Feb_A, mean_Feb_A, "Februari", "A")

standard_deviation(Mar_A, mean_Mar_A, "Maret", "A")

standard_deviation(Apr_A, mean_Apr_A, "April", "A")

standard_deviation(Mei_A, mean_Mei_A, "Mei", "A")

print()

#Mencari Standard Deviasi penjualan B (Januari - Mei)
standard_deviation(Jan_B, mean_Jan_B, "Januari", "B")

standard_deviation(Feb_B, mean_Feb_B, "Februari", "B")

standard_deviation(Mar_B, mean_Mar_B, "Maret", "B")

standard_deviation(Apr_B, mean_Apr_B, "April", "B")

standard_deviation(Mei_B, mean_Mei_B, "Mei", "B")
```

SKEWNESS

```python
from scipy.stats import skew
def skewness(data, bulan, penjualan):
    skewness_value = skew(data)
    print(f"Skewness penjualan {penjualan} {bulan} = {skewness_value:.2f}")

#Mencari Skewness penjualan A (Januari - Mei)
skewness(Jan_A, "Januari", "A")

skewness(Feb_A, "Februari", "A")

skewness(Mar_A, "Maret", "A")

skewness(Apr_A, "April", "A")

skewness(Mei_A, "Mei", "A")

print()

#Mencari Skewness penjualan B (Januari - Mei)
skewness(Jan_B, "Januari", "B")

skewness(Feb_B, "Februari", "B")

skewness(Mar_B, "Maret", "B")

skewness(Apr_B, "April", "B")

skewness(Mei_B, "Mei", "B")
```
